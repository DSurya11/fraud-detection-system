```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Class'].value_counts()
```

```python
sns.countplot(x="Class", data=df)
plt.title("Fraud (1) vs Normal (0)")
plt.show()
```

```python
sns.histplot(df['Amount'], bins=100, kde=True)
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x="Class", y="Amount", data=df)
plt.title("Fraud vs Normal Transaction Amounts")
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.hist(df['Time'], bins=50)
plt.title("Transactions Over Time")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.histplot(df[df['Class']==1]['Time'], bins=50, color="red", label="Fraud")

plt.legend()
plt.title("Frauds vs Normal Transactions Over Time")
plt.show()
```

```python
plt.figure(figsize=(10,4))

sns.histplot(df[df['Class']==0]['Time'], bins=50, color="green", alpha=0.5, label="Normal")
plt.legend()
plt.title("Frauds vs Normal Transactions Over Time")
plt.show()
```

```python
corr = df.corr()['Class'].sort_values(ascending=False)
print(corr.head(10))
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(), cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```
